```python
from gtts import gTTS
import os
```

```python
def text_to_speech(text, output_file=None):
    try:
        tts = gTTS(text, lang='fr')
        if output_file:
            tts.save(output_file)  # Sauvegarde l'audio dans un fichier
            print(f"Le fichier audio a été sauvegardé sous : {output_file}")
        else:
            os.system("mpg321 audio.mp3")  # Joue l'audio directement
    except Exception as e:
        print(f"Une erreur est survenue : {e}")

# Entrée utilisateur
text = "coup du marteau gbin ! gbin ! gbin !"
save_option = "y"

if save_option == 'y':
    file_name = "texte_en_audio.mp3"
    text_to_speech(text, file_name)
else:
    text_to_speech(text)
```
